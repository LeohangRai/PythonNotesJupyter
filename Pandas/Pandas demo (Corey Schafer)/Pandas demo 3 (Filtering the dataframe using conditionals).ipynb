## Filtering a column but the condition is True for multiple values
### isin() method

<br>
For example: if we want to filter out all the rows which have the country(column) as, not just the US but a list of countries<br>
i.e. ['US, 'India', 'UK', 'Canada'] <br>
<br>
We can accomplish by using the isin() method on the list of countries. <br>
First, we create a list of countries (The values to be True) <br>
countries = [['US, 'India', 'UK', 'Canada']] <br><br>
Then, we apply the isin() method by passing the list in this method as:
df['Country'].isin(countries)

```python
#Let's filter by countries
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = df['Country'].isin(countries)
type(filt)
```

```python
df.loc[filt, ['Country']].head()
```

## String methods within pandas
Example:<br>
str.contains()


#### Let's filter the rows where the value for LanguageWorkedWith column has the string "Python" in it.


```python
df['LanguageWorkedWith']
```

#### Here we can see that the column has values with multiple language names separaged by a semi-colon

Let's create a filter that filters the rows where the value of 'LanguageWorkedWith' column contains 'Python'

```python
has_python = df['LanguageWorkedWith'].str.contains('Python', na = False)
#Setting na = False will make sure that the Nan values are ignored or sth
type(has_python)
```

```python
has_python.value_counts()
```

```python
df.loc[has_python, 'LanguageWorkedWith']
```
