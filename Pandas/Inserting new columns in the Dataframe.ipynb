## Inserting new column into the dataframe

<PRE>
Syntax:
frame['new_column'] = [List, Series or sth like that]

### Important Note!!

The column returned from indexing a DataFrame is a view on the
underlying data, not a copy. Thus, any in-place modifications to the
Series will be reflected in the DataFrame. The column can be
explicitly copied with the Seriesâ€™s copy method.

```python
import pandas as pd
```

```python
data = {
    'state': ['Ohio', 'Ohio', 'Ohio', 'Nevada', 'Nevada', 'Nevada'],
    'year': [2000, 2001, 2002, 2001, 2002, 2003],
    'pop': [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]
    }

frame = pd.DataFrame(data)
```

```python
#Inserting new column by passing a list
frame['new random column'] = range(6)
frame
```

```python
#Inserting new column with a filter
frame['Ohio or not?']= frame.state == 'Ohio'
frame
```

 The columns in a dataframe can be deleted by using the del method
 

```python
del frame['new random column']
frame
```

```python
#Inserting new column by calculation on previous 2 columns
frame['calc'] = frame['year']/frame['pop']
frame
```
