```python
from matplotlib import pyplot as plt

import numpy as np

import math

x = np.arange(0,2*math.pi, 0.05)
y = np.sin(x)

plt.plot(x,y)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("Sine wave")
plt.show()
```

```python
from matplotlib import pyplot as plt

import numpy as np

import math

x = np.arange(0,2*math.pi, 0.05)
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y)
plt.plot(x,z)
plt.xlabel("angle")
plt.ylabel("sine")
plt.title("Sine wave")
plt.show()
```
