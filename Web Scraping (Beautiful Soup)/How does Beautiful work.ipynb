#### Read the comments (They have important info)

```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup

url = 'https://www.w3schools.com/python/ref_file_read.asp'
html = urllib.request.urlopen(url).read() #read() method returns the specified number of bytes from the file. Default is -1 which means the whole file.

#The BeautifulSoup obj is created using the whole file and parsing it using the method spedified in the second argument
soup = BeautifulSoup(html, 'html.parser') 

#In the soup object, a cleaned version of the file, (tree structre (of tags I guess)) is stored

tags = soup('a')  #tags will store a list of all the 'a' tags from the html
for tag in tags:
    print(tag.get('href', None))   #get the href value or none (None means return the whole tag value, I guess)
```

### How does Beautiful soup work
<PRE>
BeautifulSoup is smart enough to convert all this stuff so it just takes the whole file and uses this HTML parser.
There's some other options that you could have here, but that's the one that I recommend in this case.
And then you get back an object. And what happens is, in this soup variable,
it's somehow taken all the nasty bits that come from the web page
and cleaned them up, and made a little pretty tree of things.
You don't have to worry about the internal structure, it is a soup object.
