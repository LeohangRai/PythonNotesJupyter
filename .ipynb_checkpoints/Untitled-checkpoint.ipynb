```python
def fibbo(n):
    x = 1
    y = 1
    lis = [1, 1]
    
    while len(lis) < n:
        x, y = y , x+y
        lis.append(y)
        
    return lis
        
fibbo(5)
fibbo(11)
```

```python
import numpy as np

def find_duplicate(array):
    duplicate = [x for x in range(0,len(array)) if array.count(x) > 1]
    return duplicate

arr = np.array([1,22,22,34,2,3])
find_duplicate(arr)
```
